### Train a Logistic Regression Model
- same action as linear regression but compresses input values between the range of 0 and 1
- fetch_openml - The function access the open source machine learning datasets from www.openml.org, a large repository hosted by two Dutch universities; TU Eindhoven, Leiden University.
- identify handwriting
- https://security.kiwi/docs/training-regression-models/  

Get data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

Load the data via fetch_openml

```python
# MNIST dataset
mnist = fetch_openml('mnist_784')
```

Split the dataset intro training and test datasets (20k/10k)

```python
train_img, test_img, train_lbl, test_lbl = train_test_split(
    mnist.data, mnist.target, train_size=20000, test_size=10000, random_state=0
)
```

View data

```python
plt.figure(figsize=(15,4))
for index, (image, label) in enumerate(zip(train_img[0:10], train_lbl[0:10])):
    plt.subplot(1, 10, index + 1)
    plt.imshow(np.reshape(image, (28,28)), cmap=plt.cm.gray)
    plt.title('Training: {}'.format(label), fontsize = 10)
```

Train the model
- assign function to variable lr
- set a specific solver

```python
lr = LogisticRegression(solver='saga', tol=0.1)

lr.fit(train_img, train_lbl)
```

View results

```python
lr.predict(test_img[0:10])
```

Check accuracy - predictions vs. test data

```python
score = lr.score(test_img, test_lbl)
score
```
